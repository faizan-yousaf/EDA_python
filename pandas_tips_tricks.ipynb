# **14- Splitting a string into multiple columns:** 

```python
df = pd.DataFrame({"name": ["Faizan Yousaf", "Zahoor Ahmad", "Waqar Akram"], 
                   "age": [23, 24, 25],
                   "location" : ["Khanewal, Pakistan", "Faisalabad, Pakistan", "Multan, Pakistan"]})

df
```

```python
# splitting the name column into two columns: 
df.name.str.split(" ", expand = True)
```

```python
# split name into first_name and second_name 
df[['first_name', 'second_name']] = df.name.str.split(" ", expand = True)
df
```

```python
df = df.drop(columns= "second_name")
df
```

```python
# now splitting the location into two columns: city and country

df.location.str.split(", ", expand=True)
```

```python
#adding city and country columns to the dataframe:
df[["city", "country"]] = df.location.str.split(", ", expand=True)
df
```

```python
# refine data Manipulation:

df = df[["first_name", 'last_name', 'age', 'city', 'country']]
df
```

# **21- reshaping a dataframe:**

```python
fasla = pd.DataFrame([["12345", 100,200,300], ["34567", 400,500,600], ["67890", 700,800,900 ]], columns = ["zip", "factory", "warehouse", "retail"])

fasla.head()
```

```python
#wide to large format:

fasla.head().T
```

```python
fasla2 = pd.DataFrame([[1, 12345, 'factory' ], [2, "34567", "warehouse"]],
                            columns = ["user_id", "zip", "location_type"])

fasla2
```

```python
fasla
```

```python
# id_vars is the column that you want to keep as it is:
# var_name is the column that you want to melt:
# value_name is the column that you want to keep as it is:

fasla_long = fasla.melt(id_vars= 'zip', var_name= 'location_type', value_name= 'distance')
fasla_long

# both are correct:


# fasla3 = pd.melt(fasla, id_vars = 'zip', var_name = 'location_type', value_name = 'distance')
# fasla3
```
